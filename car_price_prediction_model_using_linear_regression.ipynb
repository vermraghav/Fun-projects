# Predicting prices for test data and comparing with actual values

```python
pred_price=np.exp(yhat_test)
```

```python
actual_price=np.exp(y_test)
```

```python
final_table=pd.DataFrame(pred_price,columns=['Prediction'])
```

```python
final_table
```

```python
actual_price=actual_price.reset_index(drop=True)
```

```python
final_table['Target']=actual_price
```

```python
final_table['Residual']=final_table['Target']-final_table['Prediction']
final_table['Difference%']=np.absolute(final_table['Residual']/final_table['Target']*100)
```

```python
final_table=final_table.drop(['Difference'],axis=1)
```
